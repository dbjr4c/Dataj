```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

mu_palette = ["#000000", "#e74c3c"]
race_palette = ["#1b9e77", "#d95f02"]
```

```python
data = pd.read_csv('campus_crime.csv')

# Example of some printed data, to verify the import worked
print data
```

```python
data["total_count"] = data["murder"] + data["manslaughter"] + data["forcible"]\
  + data["nonforce"] + data["robbery"] + data["agg_aslt"] + data["burglary"]\
  + data['vehicle'] + data["arson"]


data[:5]
```

```python
data.sort('total_count', ascending=False)[:20]
```

```python
data.sort('burglary', ascending=False)[:20]
```

```python
# Highlight Pinellas with facetgrid
g = sns.FacetGrid(data,
                  hue='mu_flag', 
                  palette=mu_palette,
                  size=7)

g.map(plt.scatter, "total_count", "total_enr",
      alpha=.8)
```

```python
smaller_data = data[data.total_enr < 50000]
g = sns.FacetGrid(smaller_data,
                  hue='mu_flag', 
                  palette=mu_palette,
                  size=7)

g.map(plt.scatter, "total_count", "total_enr",
      alpha=.8)
```

```python
data[(data.total_enr >= 10000) & (data.total_count == 0)].sort('total_count')
```
